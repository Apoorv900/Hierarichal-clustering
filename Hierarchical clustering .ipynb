```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:, [3, 4]].values
```

```python
import scipy.cluster.hierarchy as sch
dendrogram=sch.dendrogram(sch.linkage(X,method='ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show
```

```python
from sklearn.cluster import AgglomerativeClustering
hierarical_clustering=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
y_hierarical_clustering=hierarical_clustering.fit_predict(X)
print(y_hierarical_clustering)
```

```python
for i in range(5):
    plt.scatter(X[y_hierarical_clustering==i,0],X[y_hierarical_clustering==i,1],s=100
                ,c=['r','b','g','y','c'][i],label="cluster"+str(i+1))
plt.title("hierarical clustering graph")
plt.legend()
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```
